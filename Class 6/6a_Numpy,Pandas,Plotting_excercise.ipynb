# Python Exercises - Numpy, Pandas & Plotting

## Setup: import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1


Create a pandas dataframe that contains the following columns:

Column Name | Values
--- | ---
A | 500 random numbers drawn from a normal distribution with mean 15 and standard deviation 5
B | 500 equally spaced numbers between 1 and 50 (inclusive)
C | 500 random integers drawn between 10 and 30 (inclusive)



```python

np.random.seed(42)
column_a = np.random.normal(loc=15, scale=5, size=500) #mean/devation/amount of numbers
column_b = np.linspace(1, 50, 500)
column_c = np.random.randint(10, 31, size=500)

df = pd.DataFrame({
    'A': column_a,
    'B': column_b,
    'C': column_c
})

print(df)
```

## Problem 2


Calculate the mean of column A.

```python
df.A.mean()
```

## Problem 3


Plot a histogram of column A.

Plot a vertical line at the mean. Make the figure size (10,5). Make sure to add axis labels and a legend for the mean.




```python
plt.figure(figsize=(10, 5))
plt.hist(df['A'], bins=30, alpha=0.7, label='Data')
plt.axvline(df['A'].mean(), color='red', linestyle='--', label='Mean')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Column A with Mean')
plt.legend()
```

## Problem 4


Create a line plot of the first 100 values in column C.

Make sure to add axis labels.

```python
first_100 = df['C'].iloc[:100]
plt.plot(first_100_C)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of the First 100 Values in Column C')
```

## Problem 5


Create a new column, `D`, that contains 3 times the values in column `B` plus some random noise (`np.random.normal(0,10,500`).

---



```python
noise = np.random.normal(0, 10, size=500)
column_d = 3 * df['B'] + noise


print(column_d)
```

## Problem 6


Create a scatter plot with column `B` on the x-axis and column `D` on the y-axis.

> Indented block



Make sure to add axis labels.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['B'], df['D'], alpha=0.5)
plt.xlabel('B')  # X
plt.ylabel('D')  #Y
plt.title('Scatter Plot of Column B vs Column D')  #Title
plt.grid(True)  # Grids
plt.show()  # Display the plot
```

```
arr = np.random.randint(10, size = 9)
print(arr)
```

```
# What happens if we don't copy?
arr2 = arr
arr3 = arr.copy()
print(f"arr: {arr}")
print(f"arr2: {arr2}")
print(f"arr3: {arr3}")
arr[0] = 10
print(f"arr (updated): {arr}")
print(f"arr2 (without copying): {arr2}")
print(f"arr3 (with copying): {arr3}")
```

```
mask = np.logical_and((arr % 2 == 0),(arr >= 6))
arr*mask
```
